* 상대경로는 현재 위치한 곳을 기준으로 하는 경로입니다.
* /는 루트

* ../는 현재 위치의 상위 폴더
* ../../는 두 단계 위의 상위 폴더를 의미한다.

# 여러 파일을 한번에 로드하는 방법 

```python
file_csv = glob("data/store/소상공인*.csv")
file_csv #
#*.csv csv인 모든 파일 불러오기
```

```python
pd.read_csv(file_csv[0])
```

```python
file_list = []
for file_csv_name in file_csv:
    print(file_csv_name)
    temp = pd.read_csv(file_csv_name, low_memory=False ) # temp는 file_csv에 있는 이름 하나씩 임시로 저장
    file_list.append(temp) # file_list라는 빈 리스트에 이름들 하나씩 추가
```

```python
file_list[-1] #충북에 있는 소상공인
```

```python
pd.concat(file_list) #csv 파일을 칼럼명에 맞게 전부 합쳐줌
#컬럼명이 같다면 하나의 데이터프레임으로 합쳐주게됨
```

```python
len(file_list)
```

```python
df = pd.concat(file_list)
df.shape
```
